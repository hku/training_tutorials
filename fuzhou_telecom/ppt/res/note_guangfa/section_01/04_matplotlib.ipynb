pyplot, pylab(deprecated) 是对matplotlib 的封装，提供类似于matlab的绘图体验

### examples

```python
# example 1: plot line

import numpy as np
from matplotlib import pyplot as plt
x = np.linspace(0, 2*np.pi, 20)
y = np.sin(x)
plt.plot(x,y, 'r-.>')
plt.show()
```

```python
#example 2: scatter

import numpy as np
from matplotlib import pyplot as plt
x = np.random.rand(200)
y = np.random.rand(200)
size = np.random.rand(200)*30
color = np.random.rand(200)
plt.scatter(x, y, size, color)
plt.show()
```

```python
# example 3: histgram

import numpy as np
# from matplotlib.pylab import hist
from matplotlib import pyplot as plt
plt.hist(np.random.randn(1000), 100)
plt.show()
```

```python
# example 4: pie

import numpy as np
from matplotlib import pyplot as plt 
fig = plt.figure()
ax = fig.gca()
plt.pie([15, 30, 45, 10], labels=["A", "B", "C", "D"], startangle=90)
ax.axis('equal')
plt.show()
```

```python
# example 5: plot 3d line

import numpy as np
from matplotlib.pylab import figure
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection='3d')
theta = np.linspace(-4 * np.pi, 4 * np.pi, 100)
z = np.linspace(-2, 2, 100)
r = z**2 + 1
x = r * np.sin(theta)
y = r * np.cos(theta)
ax.plot(x, y, z)
```

```python
# example 6: plot 3d surface
import numpy as np
from matplotlib.pylab import figure
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

fig = figure()
ax = fig.gca(projection='3d')
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
xx, yy = np.meshgrid(x, y)
zz = xx**2 + yy**2
surf = ax.plot_surface(xx,yy,zz, cmap=cm.coolwarm)
fig.colorbar(surf)
```
